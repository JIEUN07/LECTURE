```python
X1 = [2,4,6,8]
X2 = [0,4,2,3]
Y = [81,93,91,97]
```

```python
# 3d 그래프 그리기
ax = plt.axes(projection ='3d')
ax.set_xlabel('Study times')
ax.set_ylabel('Private_Class')
ax.set_zlabel('Score')
ax.scatter(X1,X2,Y)
plt.show()
```

```python
# 넘파이 어레이형태로 바꾸기
X1_data = np.array(X1)
X2_data = np.array(X2)
Y_data = np.array(Y)
```

```python
# 기울기, 절편 초기화
a1 = 0
a2 = 0
b = 0

# 학습률 0.03
# 반복률 1000
lr = 0.03
epochs = 1000
```

```python
# Y값 예측, 오차 출력
for i in range(epochs):
    pre_Y = (a1 * X1_data)+(a2 * X2_data)+b
    err = Y_data - pre_Y
    # 편미분
    # 오차 함수를 a1으로 미분
    diff_a1 = -(2/len(X1)) * sum(err*X1_data)
    diff_a2 = -(2/len(X2)) * sum(err*X2_data)
    diff_b = -(2/len(Y)) * sum(err) 
    
    # 학습률 곱해 기존 a1 값 업데이트
    a1 -= lr * diff_a1
    a2 -= lr * diff_a2
    b -= lr *diff_b
    print(a1, a2, b)
    
    if i % 100 == 0:
        print('Epoch = %.04d, a1 = %.04f, a2 = %.04f, b = %.04f' %(i, a1, a2, b))
print('y = %.04fx1 + %.04fx2 + %.04f' %(a1, a2, b))
```
