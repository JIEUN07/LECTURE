### 정렬

- 1차원
```
np.sort(넘파이 배열) -> 오름차순
넘파이배열[::-1] -> 내림차순
```

- 2차원
```
np.sort(넘파이 배열, axis=0/1) -> 오름차순 
np.sort(넘파이 배열, axis=1)[::,::-1] -> 가로방향 내림차순
np.sort(넘파이 배열, axis=0)[::-1,::] -> 세로방향 내림차순
```

- np.argsort()
```
np.argsort(넘파이배열, axis=1/0) 인덱스 정렬
np.argsmax(넘파이배열, axis=1/0) 최대값 인덱스
np.argsmin(넘파이배열, axis=1/0) 최소값 인덱스
```
인덱스란 행의 순서 0~n까지

# 조건 필터링
- 비교나 논리연산자를 이용해서 데이터 추출
- Boolean Index 
```
: 배열[조건식]
: 논리연산자 ~ | &
```
- np.where(조건식)
```인덱스로 반환```

````python
arr1 = np.random.randint(1,100,10)
arr1
````

````python
# 30 이하의 숫자만 필터링
arr1 <= 30
````

````python
arr1[arr1<=30] # 트루만 반환
````

````python
# 50~70 사이의 숫자만 필터링
(arr1 >= 50) & (arr1 <= 70)
````

````python
# 50~70 사이의 숫자를 제외한 필터링
(arr1 >= 50) | (arr1 <= 70)
````

````python
# 20 보다 작거나 70보다 큰 숫자 제외
~((arr1 <= 20) | (arr1 >= 70))
````

````python
arr1[~((arr1 <= 20) | (arr1 >= 70))]
````

### np.where()

````python
arr2 = np.random.randint(1,100,25)
arr2
````

````python
# 70보다 크다
np.where (arr2>70)
````

````python
# 배열 실제값 반환 => 배열[[np.where(조건식)]]
arr2[np.where(arr2>70)]
````

````python
# 90 보다 크거나 10보다 작은 데이터만 출력하여라
np.where((arr2 > 90)| (arr2 <10))
````

````python
arr2[np.where((arr2 > 90)| (arr2 <10))]
````
