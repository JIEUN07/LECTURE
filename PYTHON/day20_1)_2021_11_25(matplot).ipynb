```python
x = np.random.randn(30)
y = np.random.randn(30)
size = np.random.rand(30)*1000
color = np.random.rand(30)

plt.scatter(x,y,
            s=size, 
            alpha=0.5,
            cmap = 'viridis', 
            c= color)
plt.colorbar()
plt.show()

plt.scatter(x,y,
            s=size, 
            alpha=0.5,
            cmap = 'Blues', 
            c= color)
plt.colorbar()
plt.show()

plt.scatter(x,y,
            s=size, 
            alpha=0.5,
            cmap = 'inferno', 
            c= color)
plt.colorbar()
plt.show()
```

```python
# 같은 결과 (for문)
x = np.random.randn(30)
y = np.random.randn(30)
size = np.random.rand(30)*1000
color = np.random.rand(30)

for colormap in ['viridis', 'Blues', 'inferno']:
  print(colormap)
  plt.scatter(x,y,
            s=size, 
            alpha=0.5,
            cmap = colormap, 
            c= color)
  plt.colorbar()
  plt.show()
```

```python
x = np.linspace(0, 10, 100)
```

```python
# 그래프 이름 추가
fig = plt.figure()
fig.suptitle("Sine and Cosine, 사인 코싸인")
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

```python
# x, y축 이름, 범례 추가
fig = plt.figure()
fig.suptitle("Sine and Cosine, 사인 코싸인")
plt.plot(x, np.sin(x), label = 'sin')
plt.plot(x, np.cos(x), label = 'cos')
plt.xlabel("x label")
plt.ylabel("y label")
plt.legend()

plt.show()
```

```python
# 범례에 프레임 추가
fig = plt.figure()
fig.suptitle("Sine and Cosine, 사인 코싸인")
plt.plot(x, np.sin(x), label = 'sin')
plt.plot(x, np.cos(x), label = 'cos')
plt.xlabel("x label")
plt.ylabel("y label")
plt.legend(frameon = True)

plt.show()
```

```python
# 범례 위치 설정
# 디폴트값은 그래프와 겹치지 않는 곳에 자동 배치

fig = plt.figure()
fig.suptitle("Sine and Cosine, 사인 코싸인")
plt.plot(x, np.sin(x), label = 'sin')
plt.plot(x, np.cos(x), label = 'cos')
plt.xlabel("x label")
plt.ylabel("y label")
plt.legend(frameon=True, loc="upper left")

plt.show()
```

```python
for loc in ["upper left", 
            "upper right", 
            "center left",
            "center right",
            "lower left", 
            "lower right", 
            "center", 
            "right",  
            "best"]:
  fig = plt.figure()
  fig.suptitle(loc)
  plt.plot(x, np.sin(x), label="sin")
  plt.plot(x, np.cos(x), label="cos")
  plt.xlabel("x label")
  plt.ylabel("y label")
  plt.legend(frameon=True, loc=loc)

  plt.show()  
```

# 여러개 그리기
- plt.figure()
- ax1 = fig.add_subplot(행, 열, 번째)

```python
x = np.linspace(0, 10, 100)
```

```python
# 1행 2열에 그래프 두개 나타내기
fig = plt.figure()

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(np.sin(x))

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(np.cos(x))

plt.show()
```

```python
# 2행 1열에 그래프 두개 나타내기
fig = plt.figure()

ax1 = fig.add_subplot(211)
ax1.plot(np.sin(x))

ax2 = fig.add_subplot(211)
ax2.plot(np.cos(x))

plt.show()
```

```python
# 2행 2열에 그래프 두개 나타내기
fig = plt.figure()

ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(np.sin(x))

ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(np.cos(x))

ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(np.cos(x))

ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(np.sin(x))

plt.show()
```

```python
# 2행 2열에 그래프 두개 나타내기
# 격자 제거
fig = plt.figure()

ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(np.sin(x))

ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(np.cos(x))
ax2.grid(False)

ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(np.cos(x))
ax3.grid(False)

ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(np.sin(x))

plt.show()
```

```python
# 2행 2열에 그래프 두개 나타내기
# 격자 없애기
# 제목 추가하기
fig = plt.figure()
fig.suptitle("sin and cos")

ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(np.sin(x))
ax1.set_title("sin")

ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(np.cos(x))
ax2.set_title("cos")
ax2.grid(False)

ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(np.cos(x))
ax3.set_title("cos")
ax3.grid(False)

ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(np.sin(x))
ax4.set_title("sin")

# 겹침 방지
plt.tight_layout()
plt.show()
```

```python
# 2행 2열에 그래프 두개 나타내기
# 격자 없애기
# 제목 추가하기
# 고양이 이미지 삽입
fig = plt.figure()
fig.suptitle("sin and cos")

ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(np.sin(x))
ax1.set_title("sin")

ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(np.cos(x))
ax2.set_title("cos")
ax2.grid(False)

ax3 = fig.add_subplot(2, 2, 3)
ax3.imshow(img)
ax3.set_title("cat")
ax3.grid(False)

ax4 = fig.add_subplot(2, 2, 4)
ax4.imshow(a)
ax4.set_title("numpy")

# 겹침 방지
plt.tight_layout()
plt.show()
```

```python
# 2행 2열에서 1행 그래프는 한개만 나타내기
# 격자 없애기
# 제목 추가하기
# 고양이 이미지 삽입
# ticks 없애기
fig = plt.figure()
fig.suptitle("sin and cos")

ax1 = fig.add_subplot(211)
ax1.plot(np.sin(x))
ax1.set_title("sin")

ax2 = fig.add_subplot(211)
ax2.plot(np.cos(x))

ax3 = fig.add_subplot(223)
ax3.imshow(img)
ax3.set_title("cat")
ax3.axis('off')

ax4 = fig.add_subplot(224)
ax4.imshow(a)
ax4.grid(True)
ax4.axis('off')

# 겹침 방지
plt.tight_layout()
plt.show()
```
