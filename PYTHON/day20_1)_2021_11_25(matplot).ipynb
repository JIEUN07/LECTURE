```python
x = np.random.randn(30)
y = np.random.randn(30)
size = np.random.rand(30)*1000
color = np.random.rand(30)

plt.scatter(x,y,
            s=size, 
            alpha=0.5,
            cmap = 'viridis', 
            c= color)
plt.colorbar()
plt.show()

plt.scatter(x,y,
            s=size, 
            alpha=0.5,
            cmap = 'Blues', 
            c= color)
plt.colorbar()
plt.show()

plt.scatter(x,y,
            s=size, 
            alpha=0.5,
            cmap = 'inferno', 
            c= color)
plt.colorbar()
plt.show()
```

```python
#   (for)
x = np.random.randn(30)
y = np.random.randn(30)
size = np.random.rand(30)*1000
color = np.random.rand(30)

for colormap in ['viridis', 'Blues', 'inferno']:
  print(colormap)
  plt.scatter(x,y,
            s=size, 
            alpha=0.5,
            cmap = colormap, 
            c= color)
  plt.colorbar()
  plt.show()
```

```python
# 2 2   
#  
#  
fig = plt.figure()
fig.suptitle("sin and cos")

ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(np.sin(x))
ax1.set_title("sin")

ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(np.cos(x))
ax2.set_title("cos")
ax2.grid(False)

ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(np.cos(x))
ax3.set_title("cos")
ax3.grid(False)

ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(np.sin(x))
ax4.set_title("sin")

#  
plt.tight_layout()
plt.show()
```
