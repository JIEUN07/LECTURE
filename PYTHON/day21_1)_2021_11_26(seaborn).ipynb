```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import numpy as np

titanic = sns.load_dataset("titanic")
```

# violinplot()

```python
sns.set_theme(style="whitegrid")
```

```python
tips = sns.load_dataset("tips")
tips
```

boxplot과 비교

```python
sns.boxplot(y=tips.tip)
plt.show()
```

```python
sns.violinplot(y=tips.tip)
plt.show()
```

```python
plt.subplot(121)
sns.boxplot(y=tips.tip)

plt.subplot(122)
sns.violinplot(y=tips.tip)

plt.show()
```

```python
plt.subplot(121)
sns.boxplot(data=tips, x="day", y="total_bill")

plt.subplot(122)
sns.violinplot(data=tips, x="day", y="total_bill")

plt.tight_layout()  # 겹침방지
plt.show()
```

```python
plt.subplot(121)
sns.boxplot(data=tips, x="time", y="total_bill")

plt.subplot(122)
sns.violinplot(data=tips, x="time", y="total_bill")

plt.tight_layout()  # 겹침방지
plt.show()
```

```python
# 데이터 순서 바꾸기
plt.subplot(121)
sns.boxplot(data=tips, x="time", y="total_bill", order = ["Dinner","Lunch"])

plt.subplot(122)
sns.violinplot(data=tips, x="time", y="total_bill", order = ["Dinner","Lunch"])

plt.tight_layout()  # 겹침방지
plt.show()
```

```python
# 가로세로 바꾸기
plt.subplot(121)
sns.violinplot(data=tips, y="time", x="total_bill", order = ["Dinner","Lunch"])

plt.subplot(122)
sns.violinplot(data=tips, x="time", y="total_bill", order = ["Dinner","Lunch"])

plt.tight_layout()  # 겹침방지
plt.show()
```

```python
# 날짜별로 흡연자 비흡연자 나누기
plt.figure(figsize=(10,4))
plt.subplot(121)
sns.violinplot(data=tips, x="day", y="total_bill", hue="smoker")

plt.subplot(122)
sns.violinplot(data=tips, x="day", y="total_bill", hue="smoker")

plt.tight_layout()  # 겹침방지
plt.show()
```

```python
# 반반보기 split="True"
plt.figure(figsize=(10,4))
plt.subplot(121)
sns.violinplot(data=tips, x="day", y="total_bill", hue="smoker")

plt.subplot(122)
sns.violinplot(data=tips, x="day", y="total_bill", hue="smoker", split="True")

plt.tight_layout()  # 겹침방지
plt.show()
```

```python
# 디폴트 inner = "box"
# box, quantile, point, stick

plt.figure(figsize=(10,4))
plt.subplot(121)
sns.violinplot(data=tips, x="day", y="total_bill")

plt.subplot(122)
sns.violinplot(data=tips, x="day", y="total_bill", inner = "quartile")

plt.tight_layout()  # 겹침방지
plt.show()
```

# stripplot()

```python
plt.figure(figsize=((10,4)))
plt.subplot(121)
sns.violinplot(data=tips, x="day", y="total_bill")

plt.subplot(122)
sns.stripplot(data=tips, x="day", y="total_bill")

plt.tight_layout()
plt.show()
```

겹쳐그리기

```python
plt.figure(figsize=((10,4)))
plt.subplot(121)
sns.violinplot(data=tips, x="day", y="total_bill")
sns.stripplot(data=tips, x="day", y="total_bill", color="0.5")

plt.subplot(122)
sns.stripplot(data=tips, x="day", y="total_bill")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=((10,4)))
plt.subplot(121)
sns.boxplot(data=tips, x="day", y="total_bill")
sns.stripplot(data=tips, x="day", y="total_bill", color="0.5")

plt.subplot(122)
sns.stripplot(data=tips, x="day", y="total_bill")

plt.tight_layout()
plt.show()
```

# np.random에서 시드고정하기
- np.random.seed(42)
- 42는 고정시키는 목적 (별다른 뜻은 없음)

```python
# 할상 같은 값 나옴
np.random.seed(42)

a = np.random.rand(10,8)
print(a[0,0])
```

```python
# 색 범위 설정
np.random.seed(42)
a = np.random.rand(10,8)

plt.figure(figsize=((10,4)))
plt.subplot(121)
sns.heatmap(a)

plt.subplot(122)
sns.heatmap(a, vmin = 0.2, vmax=0.8)

plt.show()
```

```python
# 숫자 표기
plt.figure(figsize=((10,4)))
plt.subplot(121)
sns.heatmap(a)

plt.subplot(122)
sns.heatmap(a, annot=True)

plt.show()
```

```python
# 중간에 사이 띄워주기
plt.figure(figsize=((10,4)))
plt.subplot(121)
sns.heatmap(a)

plt.subplot(122)
sns.heatmap(a, linewidth=0.5)

plt.show()
```

```python
# 색 테마 바꾸기
plt.figure(figsize=((10,4)))
plt.subplot(121)
sns.heatmap(a)

plt.subplot(122)
sns.heatmap(a, linewidth=0.5, cmap="Blues")

plt.show()
```

```python
# 컬러바 제거
# 디폴트는 cbar=True
plt.figure(figsize=((10,4)))
plt.subplot(121)
sns.heatmap(a)

plt.subplot(122)
sns.heatmap(a, linewidth=0.5, cmap="Blues", cbar=False)

plt.show()
```
