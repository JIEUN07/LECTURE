```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
# from tensorflow import keras
```

# 흑백 영상 자르기

## 데이터 준비

```python
(train_x, train_y), (test_x, test_y) = tf.keras.datasets.mnist.load_data()
```

```python
image = train_x[0]
plt.imshow(image, cmap="binary")
plt.show()
```

```python
print(image.shape)
```

```python
a = np.arange(12).reshape((3,4))
print(a)
```

```python
cropped = image[5:20, 5:20]
plt.imshow(cropped, cmap="binary")
plt.show()
```

## 파일로 저장하고 읽기

```python
plt.imsave("cropped_mnist.jpg",cropped)
```

```python
loaded = plt.imread("cropped_mnist.jpg")
print(loaded.shape)
plt.imshow(loaded)
plt.show()
```

```python
np.save("cropped.npy", cropped)
```

```python
loaded = np.load("cropped.npy")
print(loaded.shape)
plt.imshow(loaded, cmap="binary")
plt.show()
```

# 인물사진 자르기

## 데이터 준비

```python
image = plt.imread("iu.jpg")
print(image.shape)
plt.imshow(image)
plt.show()
```

```python
cropped = image[0:170, 25:170, :]
plt.imshow(cropped)
plt.show()
```

## 파일로 저장하고 읽기

```python
plt.imsave("iu_face.jpg", cropped)
loaded = plt.imread("iu_face.jpg")
plt.imshow(loaded)
plt.show()
```

## 이미지 리사이즈

```python
import cv2

resized = cv2.resize(cropped, dsize=(224,224))
plt.imshow(resized)
plt.show()

print(cropped.shape)
print(resized.shape)
```
