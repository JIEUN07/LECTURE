```python
# 네이버 검색 API예제는 블로그를 비롯 전문자료까지 호출방법이 동일하므로 blog검색만 대표로 예제를 올렸습니다.
# 네이버 검색 Open API 예제 - 블로그 검색
# 아이디와 시크릿정보 지정
client_id = "o9BQISjJKxXU9QTbPaxO"
client_secret = "GWo3_bi_An"

# 검색 단어 입력
encText = urllib.parse.quote("머신러닝")

n= 20
# 요청변수 display = 숫자
url = "https://openapi.naver.com/v1/search/blog?query=" + encText + "&display=" + str(n )# json 결과
# url = "https://openapi.naver.com/v1/search/blog?query=" + encText # json 결과
# url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # xml 결과
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
response = urllib.request.urlopen(request)
rescode = response.getcode()
if(rescode==200):
    response_body = response.read()
    print(response_body.decode('utf-8'))
else:
    print("Error Code:" + rescode)
```

```python
# API 정보 => JSON텍스트
json_text = response_body.decode('utf-8')
type(json_text)
```

```python
# 2) 리스트 딕셔너리 구조화
# 딕셔너리가 삽입된 리스트
# blog_list = json.loads(json_text) => 딕셔너리
blog_list = json.loads(json_text)["items"]
type(blog_list)
```

```python
blog_list
```

```python
# 리스트 딕셔너리 => 데이터프레임
df_blog = pd.DataFrame(blog_list)
df_blog
```

```python
# <b> 태그가 섞여있음
# 전처리 후 데이터프레임 생성
# 빈 데이터프레임 생성
df_blog2 = pd.DataFrame(columns=['title', 'link', 'description', 'bloggername','bloggerlink', 'postdate' ])
df_blog2
```

```python
# 특정 문자열을 삭제 => replace(old,new)
blog_list[0]['title'].replace('<b>','').replace('</b>','')
```

```python
# 데이터 프레임에 행을 툭하면서 전처리해서 아이템 삽입
idx = 0
for row in blog_list:
    title = row['title'].replace('<b>','').replace('</b>','')
    link = row['link']
    description = row['description'].replace('<b>','').replace('</b>','')
    bloggername = row['bloggername']
    bloggerlink = row['bloggerlink']
    postdate = row['postdate']
    # 행 추가 후 데이터 삽입
    df_blog2.loc[idx] = [title, link, description, bloggername, bloggerlink, postdate ]
    idx += 1
```

```python
df_blog2
```

```python
# 데이타프레임에 행을 추가하면서 전처리해서 아이템 삽입 
# 데이타프레임의 행 
idx = 0
for row in blog_list:
    title = row['title'].replace('<b>','').replace('</b>','')
    link = row['link']
    description = row['description'].replace('<b>','').replace('</b>','')
    bloggername = row['bloggername']
    bloggerlink = row['bloggerlink']
    postdate = row['postdate']
    # 행 추가후 데이타삽입 
    df_blog2.iloc[idx] = [title, link, description, bloggername, bloggerlink, postdate]
    idx += 1
```
