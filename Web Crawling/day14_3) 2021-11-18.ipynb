# 예제 1
알렉사 사이트의 50위
- https://www.alexa.com/topsites

````python
# url 요청 => response 객체생성
url = 'https://www.alexa.com/topsites/countries/KR'
res = requests.get(url)

# reponse 객체 => 텍스트화
html_str = res.text
html_str
````

````python
soup = BeautifulSoup(html_str, 'html.parser')
soup
````

````python
# 리스트로 지정
div_list = soup.find_all(class_='tr site-listing')
len(div_list)
````

````python
div_list[-1]
````

````python
div_list[-1].select_one('.td').text
````

````python
# 순위 rank
print(div_list[-1].select_one('.td').text)
# 사이트 제목 title
print(div_list[-1].find('a').text)
# 사이트 url
print(div_list[-1].find('a')['href'])
# daily_time
print(div_list[-1].find_all(class_='td right')[0].text)
# daily_pageviews
print(div_list[-1].find_all(class_='td right')[1].text)
````

````python
# 2차원 리스트 저장
temp_list = []
for row in div_list:
    rank =row.select_one('.td').text
    title = row.find('a').text
    daily_time = div_list[-1].find_all(class_='td right')[0].text
    daily_pageviews = div_list[-1].find_all(class_='td right')[1].text
    temp_list.append([rank,title, daily_time, daily_pageviews])    
````

````python
# 데이터프레임으로 저장
df_alexa = pd.DataFrame(temp_list, columns=[rank,title, daily_time, daily_pageviews])
df_alexa
````

````python
df_alexa.tail()
````

````python
# csv 파일로 저장
df_alexa.to_csv('output2/alexa.csv', index=False)
````
